- Lo mismo pero con cuatro nodos (10, 20, 30, 40)

```python
class Nodo:
    def __init__(self, value, siguiente = None):
        self.data = value
        self.next = siguiente
```

```python
head = Nodo(10)
print(head.data)           #Se obtuvo el nodo con valor 10.
```

```python
head = Nodo(10)
head.next = Nodo(20)
print(f"Primero = {head.data}")         #Se obtuvo el nodo con valor 10 y 20.
print(f"Segundo = {head.next.data}")

#Lo mismo pero ahorrando codigo
head = Nodo(10, Nodo (20))
print(f"Primero = {head.data}")
print(f"Segundo = {head.next.data}")

#Transversal

curr_node = head
while curr_node != None:
        print(f"| {curr_node.data} | -->", end="")
        curr_node = curr_node.next
print("")
```

```python
head = Nodo(10, Nodo(20, Nodo(30, Nodo(40))))
curr_node = head
while curr_node != None:
        print(f"| {curr_node.data} | -->", end="")
        curr_node = curr_node.next

print()

#Imprimir el primero
print("Primer valor", head.data)
#imprimir el ultimo
curr_node = head
while curr_node.next != None:
    curr_node = curr_node.next
print("Ultimo valor", curr_node.data)
```

```python
#Insertar valores
curr_node = head
while curr_node.data != 20: #Buscamos el nodo
    curr_node = curr_node.next
print(curr_node.data) #Corrobar que estamos en el ultimo

#   Insertar el 25 entre 20 y 30
curr_node = head
while curr_node.data != 20:
    curr_node = curr_node.next
print(curr_node.data)

#      Crear Nodo 25
curr_node.next = Nodo(25, curr_node.next )
print("Después de insertar a 25 ")
#   Trasnversal
curr_node = head
while curr_node !=  None:
    print(f"| {curr_node.data } | ---> " , end="")
    curr_node = curr_node.next
print("")
```

```python
class Nodo:
    def __init__(self, value, siguiente = None):
        self.data = value
        self.next = siguiente

class LinkedList:
    def __init__ (self):
        self.head = None
    def is_empty (self): #True or false
        return self.head == None

    def get_tail (self):
       if self.is_empty():
           return None
       else:
            curr_node = self.head
            while curr_node.next != None:
                curr_node = curr_node.next
            return curr_node

    def append (self, value):
        if self.head == None:
            self.head = Nodo (value)
        else:
            curr_node = self.head
            while curr_node.next != None:
                curr_node = curr_node.next   #Este while nos movera al final
            curr_node.next = Nodo (value)

    def transversal (self):
        curr_node = self.head
        while curr_node != None:
            print(f"| {curr_node.data} | -->", end="")
            curr_node = curr_node.next
        print("")

    def remove (self, value):
        if self.head == None:
            print("Lista vacia, no se puede remover nada")
        else: #De lo contrario:
            curr_node = self.head
            aux = None
            while curr_node.data != value: #Revisar cuando el valor no exista
                aux = curr_node
                curr_node = curr_node.next
            
            if curr_node != None:
                aux.next = curr_node.next
                curr_node = None
    
    def prepend (self,value):
        if self.head == None:
            print("ESTA VACIO, NO SE PUEDE")
        else:
            aux = self.head
            self.head = Nodo (value, aux)
        
```

```python
lista = LinkedList()
if lista.is_empty():
    print("La lista esta vacia")
else:
    print("La lista no esta vacia")
    
lista.append(10)
lista.append(20)
lista.append(30)
lista.append(40)
lista.append(50)

nodo = lista.get_tail()
if nodo != None:
    print(nodo.data)
else:
    print("Lista vacia")

lista.transversal()
lista.remove(30)
lista.transversal()
lista.prepend(5)
print("Despues del preppend")
lista.transversal()
```
